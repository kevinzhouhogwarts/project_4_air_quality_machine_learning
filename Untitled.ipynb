```python
import pandas as pd
import requests
from config import api_key, email
```

```python
# API endpoint with your credentials
url = f"https://aqs.epa.gov/data/api/dailyData/byState?email={email}&key={api_key}&param=44201&bdate=20230101&edate=20230131&state=06"

# Send the request to the API
response = requests.get(url)
```

```python
# Check if the request was successful
if response.status_code == 200:
    # Parse JSON response
    data = response.json()
    # Load data into a Pandas DataFrame
    df = pd.json_normalize(data['Data'])
    print(df.head())
else:
    print(f"Error: {response.status_code}")
```
