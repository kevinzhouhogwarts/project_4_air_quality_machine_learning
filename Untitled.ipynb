```python
import pandas as pd
import requests
from config import api_key, email
```

```python
# API endpoint with your credentials
url = f"https://aqs.epa.gov/data/api/annualData/byState?email={email}&key={api_key}&param=44201&bdate=20210101&edate=20211231&state=06"

# Send the request to the API
response = requests.get(url)
```

```python
# Check if the request was successful
if response.status_code == 200:
    # Parse JSON response
    data = response.json()
    # Load data into a Pandas DataFrame
    df = pd.json_normalize(data['Data'])
    print(df.head())
else:
    print(f"Error: {response.status_code}")
```

```python
unique_counts = df.nunique()
print(unique_counts)
```

```python
df.to_csv("filename-annual-2021.csv", index=False)
```
