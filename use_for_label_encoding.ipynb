```python
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import LabelEncoder
```

## Use this notebook to convert input csv into encoded output csv
All columns will be encoded to integer categories

```python
# Input  your filepath
filepath = "Resources/updated_tickets_2024_09_selected.csv"

# Read the csv from filepath
data = Path(filepath)
df = pd.read_csv(data)
df
```

```python
# Encode all existing columns to new categorical columns
label_encoders = {}
for column in df:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
label_encoders
```

```python
# Export df to csv
df.to_csv("Resources/encoded_tickets_2024_09_selected.csv", index=False)
```
